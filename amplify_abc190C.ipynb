```python
from amplify import gen_symbols, BinaryPoly
from amplify import sum_poly
from amplify.constraint import equal_to
import matplotlib.pyplot as plt
from amplify import Solver
from amplify.client import FixstarsClient
import matplotlib.ticker as ticker   
import os  
```

```python
def q_annealing(n, ab, k, cd):
    q = gen_symbols(BinaryPoly, n)
    q2 = gen_symbols(BinaryPoly, n, (k, 2))

    f = 0
    for i in range(m):
        f -= q[ab[i][0]-1] * q[ab[i][1]-1]
    
    # g = []
    g = 0
    h = [-q[i] for i in range(n)]
    for i in range(k):
        # g.append(equal_to(q2[i][0] + q2[i][1],1))
        
        g += 10000 * (2 * q2[i][0] * q2[i][1] - q2[i][0] - q2[i][1] + 1)
        h[cd[i][0]-1] += q2[i][0]
        h[cd[i][1]-1] += q2[i][1]
    for i in range(n):
        h[i] = 100 * h[i] ** 2
        
    f += sum_poly(h)
    
    # model = f + sum(g)
    model = f + g
        
    # 計算実行------------------------------------------------------------------------------------------
    client = FixstarsClient()
    with open("tokenfile.txt", "r") as f:
        token = f.readline()
    client.token = token
    client.parameters.timeout = 9900  # タイムアウト9秒
    solver = Solver(client)

    result = solver.solve(model)
    if len(result) == 0:
        raise RuntimeError("Any one of constraints is not satisfied.")

    energy, values = result[0].energy, result[0].values    
    # print(values)
    ans = 0
    for i in range(m):
        ans += values[ab[i][0]-1] * values[ab[i][1]-1] 
    
    return ans

    
```

```python
folder = "ABC190/C/"
files = os.listdir(folder + "in")

for file in files:
    with open(folder + "in/" + file, "r") as f:
        n, m = [int(i) for i in f.readline().split()]
        ab = [[int(i) for i in f.readline().split()] for i in range(m)]
        k = int(f.readline())
        cd = [[int(i) for i in f.readline().split()] for i in range(k)]

    with open(folder + "out/" + file, "r") as f:
        ans = int(f.readline())

    result = q_annealing(n, ab, k, cd)
    print(file, "Amplify結果 =", result, "正解 =", ans)
```
